```python
import pandas as pd

# 1. Load the CSV
df = pd.read_csv("../data/raw/Posts_nature_org_18_Apr_2024_17_Apr_2025_7883.csv", sep=';')

# 2. Show available column names for inspection
print("Columns in the original file:")
print(df.columns.tolist())

# 3. Define the exact columns you want to keep
columns_to_keep = [
    "Caption",
    "Image Link",
    "Eng. Rate by Followers",
    "Date",
    "Type"
]

# 4. Clean column names (strip whitespace, fix misnamed headers)
df.columns = df.columns.str.strip()

# 5. Retain only the selected columns
df_cleaned = df[columns_to_keep]

# 6. Save to a cleaned file
df_cleaned.to_csv("../data/cleaned/Posts_nature_org_CLEANED.csv", index=False)

print("âœ… File cleaned and saved.")
df_cleaned.head()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style = "darkgrid")
df = pd.read_csv("../data/processed/Posts_carbonbrief_18_Apr_2024_17_Apr_2025_2db6 6_CLEANED.csv")

print("Dataset shape:", df.shape)
df.head()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
sns.set(style="whitegrid")

# ----- 1. Post Type Distribution -----
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Type', order=df['Type'].value_counts().index, palette='magma')
plt.title("Distribution of Post Types")
plt.xlabel("Post Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ----- 2. Caption Lengths -----
df['Caption Length'] = df['Caption'].apply(lambda x: len(str(x).split()))

plt.figure(figsize=(8, 5))
sns.histplot(df['Caption Length'], bins=30, kde=True, color="teal")
plt.title("Caption Length Distribution")
plt.xlabel("Number of Words in Caption")
plt.ylabel("Number of Posts")
plt.tight_layout()
plt.show()

# ----- 3. Engagement Rate Distribution -----
plt.figure(figsize=(8, 5))
sns.histplot(df['Eng. Rate by Followers'], bins=30, kde=True, color="purple")
plt.title("Engagement Rate by Followers Distribution")
plt.xlabel("Engagement Rate")
plt.ylabel("Number of Posts")
plt.tight_layout()
plt.show()
```
